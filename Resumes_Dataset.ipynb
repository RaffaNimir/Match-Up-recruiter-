```python
import fitz
import pandas as pd

# create an empty dataframe to store the text
df = pd.DataFrame(columns=['filename', 'text'])

# loop through all the PDF files
for i in range(1, 31):
    # open the PDF file
    with open(f'{i}.pdf', 'rb') as f:
        # read the PDF file contents into a byte string
        pdf_data = f.read()

    # create a PDF document object
    pdf_doc = fitz.open(stream=pdf_data, filetype='pdf')

    # loop through all pages in the PDF document
    text = ''
    for page in pdf_doc:
        # extract the text from the page and append it to the string
        text += page.get_text()

    # close the PDF document
    pdf_doc.close()

    # add the filename and text to the dataframe
    df = df.append({'filename': f'{i}.pdf', 'text': text}, ignore_index=True)

# save the dataframe to a CSV file
df.to_csv('pdf_texts.csv', index=False)
```
